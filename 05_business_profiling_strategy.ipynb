# 06 – Business Profiling & Strategic Recommendations

## Objective
- Profile each customer segment
- Assign meaningful business names
- Translate patterns into marketing strategies

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("df_cleaned.csv")

metric_features = [
    'income','frq','rcn','clothes','kitchen',
    'small_appliances','toys','house_keeping',
    'per_net_purchase','spent_online'
]
```

## Cluster Profiles

```python
cluster_profiles = df.groupby('labels')[metric_features].mean().round(2)
cluster_profiles
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(
    cluster_profiles.T,
    annot=True,
    cmap='PiYG',
    center=0
)

plt.title("Cluster Feature Means")
plt.xlabel("Cluster")
plt.ylabel("Features")
plt.show()
```

##  Cluster 0 – Affluent Digital Spenders

###  Key Characteristics (Standardized Feature Means)

- **Income:** `+0.94` ↑ (High income)
- **Purchase Frequency (frq):** `+1.30` ↑ (Very frequent buyers)
- **Online Spending (spent_online):** `+2.00` ↑↑ (Extremely high digital usage)
- **Clothes Spending:** `-0.29` ↓ (Below average fashion interest)
- **Average Purchase Value (per_net_purchase):** `-0.13` ≈ (Around average basket size)

---

###  Customer Profile

- High-income customers  
- Shop frequently  
- Strong preference for online channels  
- Not strongly fashion-driven  
- Likely convenience-oriented and digitally engaged  

This segment represents digitally mature, high-value customers who interact heavily through online channels but are not driven by fashion trends.

---

###  Strategic Recommendations

**Primary Focus: Digital Optimization**

- Launch **digital-first marketing campaigns**
- Use **AI-powered personalized product recommendations**
- Promote **subscription-based services**
- Offer **exclusive online loyalty perks**
- Implement **targeted cross-sell strategies via email/app**

---

###  Business Opportunity

This segment has:
- High purchasing power

Optimizing digital touchpoints for this group can significantly increase customer lifetime value (CLV).
